```python
import pandas as pd

# 读取Excel文件
excel_file = '成績資料匯出拷貝.xlsx'
df = pd.read_excel(excel_file)

# 重命名列名并去除 `id` 列
df.columns = [
    'event_name', 'index', 'unit_name', 'name', 'gender', 'event', 'score', 'status'
]

# 检查并清理 `index` 列
df['index'] = pd.to_numeric(df['index'], errors='coerce').fillna(0).astype(int)

# 保存为CSV文件，不包括 `id` 列
csv_file = '成績資料匯出.csv'
df.to_csv(csv_file, index=False)
df
```
